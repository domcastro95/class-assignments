```python
def makeBirdDict(filename):
    # this function opens the data file from Thomas et al and extract 
    birdDict = {}
    fobj = open(filename,"r", encoding = "ISO-8859-15") #open the file
    all_records = fobj.readlines()[2:]#read the file 
    for line in all_records:
        record_elements = line.split(",")#split lines on comma , # this makes a string into a list of elements
        species = record_elements[0]
        family = record_elements[2]
        birdDict[species] = family  #stick species and family into birdDict
    
    return birdDict 
```

```python
seq = ("ATTTGTCTTC")
def get_at_content(dna, sig_figs=2):
    length = len(dna)
    a_count = dna.upper().count('A')
    t_count = dna.upper().count('T')
    at_content = (a_count + t_count) / length
    return(round(at_content, sig_figs))
```
